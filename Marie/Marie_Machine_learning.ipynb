```python
#drop columns that won't  be used
clean_df = trial_df.drop(columns = ["AVG of CALC BASE WT IN GM","AVG of MME Conversion Factor", "AVG of unit",\
                                    "Buyer City", "Buyer Name","DOS", "DosageUnit N","Number of Records", \
                                    "Reporter City", "Reporter Name","Reporter State",  "Reporter family",\
                                    "Strength n"])
clean_df.head()
```

```python
clean_df["risk"] = clean_df["SUM of QUANTITY"]>=10000
```

```python
clean_2012_df = clean_df.loc[high_risk_df["Transaction Year"]==2012]
```

```python
clean_2012_df["risk2"] = clean_df["risk"].astype(int)
clean_2012_df.head()
```

```python
#get dummies for states (make states into columns)
dummy_df = pd.get_dummies(clean_2012_df["Buyer State"])
dummy_df.head()
```

```python
#Listing the column names
#list(dummy_df.columns)
```

```python
final_df = clean_2012_df.join(dummy_df, how="left")
final_df.head()
```

```python
pre_reg_df = final_df
```

```python
reg_df = final_df.drop(columns = ["Transaction Year","Buyer State","risk", "risk2"])
reg_df.head()
```

```python
#define var for decision tree
target  = pre_reg_df["risk2"]
target_names = ["high_risk", "low_risk"]
feature_names = reg_df.columns
```

```python
#Machine learning
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(reg_df, target)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
